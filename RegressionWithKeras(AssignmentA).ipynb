```python
#Importing required libraries
import pandas as pd
import numpy as np
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Importing the dataset
df = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
df.head()
```

```python
#Importing the library for splitting the data
from sklearn.model_selection import train_test_split
```

```python
column_names = df.columns
```

```python
predictor = df[column_names[column_names!='Strength']]
target = df['Strength']
```

```python
x_train, x_test, y_train, y_test = train_test_split(predictor, target, test_size=0.3)
```

```python
#creating the model
def regression_model():
  model = Sequential()
  model.add(Dense(10, activation='relu', input_shape=(x_train.shape[1],)))
  model.add(Dense(1))

  model.compile(optimizer='adam', loss='mean_squared_error')
  return model
```

```python
#Model Evaluation
from sklearn.metrics import mean_squared_error
```

```python
mse_list=[]
```

```python
for _ in range(50):
  model = regression_model()
  model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=50, verbose=2)

  y_pred = model.predict(x_test)
  mse = mean_squared_error(y_test, y_pred)
  mse_list.append(mse)
```

```python
mean_mse = np.mean(mse_list)
mean_mse
```

```python
std_mse = np.std(mse_list)
std_mse
```
